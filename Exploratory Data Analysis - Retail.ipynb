```python
sns.pairplot(df,hue='Region')
plt.show()
```

```python
segment=(df['Segment'].value_counts()/len(df['Segment'])*100)
plt.rcParams['figure.figsize']=(12,6)
segment.plot(kind='pie')
plt.title('Bar graph of segment',)
plt.show()
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
sns.countplot(df['Category'],hue=df['Region'])
plt.title('Bar graph of Category')
plt.show()
```

```python
high_sales=df[df['Sales']>3000]
high_sales.head()
```

```python
sns.barplot(y=high_sales['State'],x=high_sales['Sales'],data=df)
plt.show()
```

```python
df['Sales'].describe()
```

```python
low_profit=df[df['Profit']<0]
low_profit
```

```python
sns.barplot(y=low_profit['Region'],x=low_profit['Profit'],data=df)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(10,5)
data=['Quantity','Discount','Profit','Sales']
sns.heatmap(df[data].corr(),annot=True,fmt='0.2f')
plt.title('Corelation b/w variables')
plt.show()
```
